```
import pandas as pd
```

```
from datetime import datetime

def parser (d):
    d = d.split('.')[0]
    try:
        return datetime.strptime(d, "%Y-%m-%dT%H:%M:%S" )
    except:
        print(d)

predictions = pd.read_csv('predictions.csv', parse_dates=['e'], date_parser=parser, index_col=0)
predictions.tail()
```

```
predictions = predictions.drop('predicted prob clinton (predictwise probabilities)', 1)
predictions['predicted market price'] = predictions['predicted market price'].map(lambda x: x*100)
```

```
predictions.plot(legend=False)
```

```
predictions.plot(legend=True, ylim=(30,100), colormap="Set2")
```
